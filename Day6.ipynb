7-Extract only even numbers from a list.by using filter and lamda.

```python
num=[1,2,3,4,5,6,7,8,9,0]
even=list(filter(lambda x:x%2==0,num))
print(even)
```

8-Double all values in a list.By using map and lambda

```python
num=[1,2,3,4,5,6,7,8,9]
a=list(map(lambda x:x*2,num))
print(a)
```
