1-Take a number and check whether it is positive, negative, or zero.


```python
num=int(input("Enter a number:"))
if num>0:
    print("Positive")
elif num<0:
    print("negative")
else:
    print("zero")
```

2-Based on marks input, print grade A/B/C/Fail.

```python
grade=int(input("Enter a grade :"))
if grade>=90:
    print("Grade A")
elif grade>=80:
    print("Grade B")
elif grade>=70:
    print("Grade C")
else :
    print("fail")
```

3-Print numbers from 1 to 20 using a loop.


```python
num=1
while num<=20:
    print(num)
    num+=1
    
```

4-Print numbers 10 → 1 using range() .

```python

for i in range(10,0,-1):
    print(i)
```
