2. Write a program that takes a user’s full name (with spaces), converts it to both uppercase and lowercase, uses strip() and split(), and then rejoins the words.

```python
nam=input("Enter namre :")
nam=nam.strip()
nam2=nam.split()
join=" ".join(nam2)
print(nam)
print(nam2)
print(join)
print(join.upper())
print(join.lower())
```
