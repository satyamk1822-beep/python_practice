1. Create a file and write 5 lines using "w" mode.

```python
with open("myfile.txt","w") as f:
    f.write("My name is Satyam\n")
    f.write("Pay  tax with smile...\n")
    f.write("I tried but they wanted money!\n")
    f.write("Don\'t feel bad....\n")
    f.write("Take out your document")
```

2. Read the file using "r" mode.

```python
with open("myfile.txt",'r') as f:
    print(f.read())
```

3. Append a new line using "a" mode.

4. Use "r+" mode to read and write in the same file.

```python
with open("new_file.txt",'r+') as f:
    print(f.read())
    f.write("\n We are using r+ statement right now")
```

5. Print cursor position using tell() .

```python
f=open("newfile.txt",'r')
print(f.tell())
data = f.read(5)
print(data)
print(f.tell())
f.close()
```

6. Move cursor using seek() and re-read content.


```python
f = open("newfile.txt", "r")

data1 = f.read(5)
print(data1)

f.seek(0)          # cursor wapas start par
data2 = f.read(5)
print(data2)

f.close()
```

7. Store student data in a JSON file.


```python
import json

student = {
    "name": "Aarav",
    "age": 20,
    "course": "BCA"
}

with open("student.json", "w") as file:
    json.dump(student, file)
```

8. Read JSON file and print student details.

```python
import json

with open("student.json", "r") as file:
    student = json.load(file)

print("Name:", student["name"])
print("Age:", student["age"])
print("Course:", student["course"])
```

10. Rename a file from old_name.txt to new_name.txt using the os
