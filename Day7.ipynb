1-Use sqrt() and factorial()

```python
import math

num=int(input("Enter a number: "))
print("Square root:",math.sqrt(num))
print("Factoeial:",math.factorial(num))

```

2-Random Module Generate random number and name

```python
import random
number = random.randint(1, 100)
print("Random number:", number)
names = ["Ram", "Shyam", "Aman", "Riya"]
print("Random name:", random.choice(names))
```

3-Time module Pause program using sleep()

```python
import time
print("Program started")
time.sleep(3)   
print("Program resumed after 3 seconds")
```

4-Measure time Measure loop execution time

```python
import time
start = time.time()
for i in range(1000000):
    pass
end = time.time()
print("Execution time:", end - start, "seconds")
```

5-Statistics module Find mean and median

```python
import statistics
data = [10, 20, 30, 40, 50]
print("Mean:", statistics.mean(data))
print("Median:", statistics.median(data))
```
